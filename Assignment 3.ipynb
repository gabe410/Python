```python
#we import the pandas library because we will be reading in datasets and creating a dataframe
import pandas as pd

#reading in the data set from my local library
#the name of the dataset is data_set
data_set = pd.read_csv('/Users/gabem/Documents/IS678 Data Analytics in Cybersecurity/book/ch05/data/county-data.csv', header=0)
```

```python
#we print the data set so that we can see the values we are working with
#the columns should be subregion, region, pop, income, ipaddr, ufo2010
#there should be 3072 rows x 6 columns
print(data_set)
```

```python
#this function will return a number of arrays that contain the data for each of the testing and training sets
from sklearn.model_selection import train_test_split 
```

```python
#In this code block, we are going to divide the data into the test set and the training set

#This function will split data into a training set and a testing set based on how we choose to split the data

#The arguments for this function will be the 3 columns we are going to use: pop, income, ufo2010

#Since we need population and income to be the basis for predicting substantial UFO apperances, we will need
#these two columns in addition to the ufo2010 column, which tells us how many appearances there are

#we want the test size to be 1/4 the size of the data_set data frame, and then we must convert it to an integer
#this is done by using int(len(data_set)/4)

#the function will return a pair of values: one belonging to train set, and one belonging to test set for 
#each column
x_train, x_test, y_train, y_test, z_train, z_test = train_test_split(data_set['pop'], data_set['income'],
                                                                    data_set['ufo2010'], 
                                                                     test_size = int(len(data_set)/4))
```

```python
#Here we will create training sets into a single data frame
#create a dictionary using lists (pop, income, and ufo2010) and their 
#corresponding training values (x_train, y_train, z_train)
#the training dataframe will be stored in the train_df value
train_dict = {"pop": x_train, "income": y_train, "ufo2010": z_train}
train_df = pd.DataFrame(train_dict)
train_df

#the output should be a table with columns pop, income, and ufo2010
#it should have 2304 rows x 3 columns
```

```python
#Here we will create the testing sets into a single data frame
#create a dictionary using lists (pop, income, and ufo2010) and their 
#corresponding testing values (x_test, y_test, z_test)
#the testing dataframe will be stored in the test_df value
test_dict = {"pop": x_test, "income": y_test, "ufo2010": z_test}
test_df = pd.DataFrame(test_dict)
test_df

#the output should be a table with columns pop, income, and ufo2010
#it should have 768 rows x 3 columns
```

```python
import statistics as stats #we will need to import the statistics library in order to calculate the mean
```

```python
#The following algorithm will calculate the mean for the population and income for regions that have
#substantial appearances and regions that do not have substantial appearances

#Substantial appearances are classified as any region that has had more than 333 UFO sightings in that
#specific region

#Here we will calculate means of population and income for regions with and without substantial UFO appearances

#Means for regions with substantial appearances
#x is a temporary initial value
#take only values from 'ufo2010' column of training table that have greater than 333 appearances (substantial)
x = train_df[train_df.ufo2010 > 333]
#calculate mean of population column for substantial appearances and store it in train_pop_Sub_mean value
train_pop_Sub_mean = stats.mean(x['pop'])
#calculate mean of income column for subs. appearances and store it in train_income_Substantial_mean value
train_income_Substantial_mean = stats.mean(x['income'])

#Means for regions with no substantial appearances
#take only values from 'ufo2010' column of training table that have less than 333 appearances (not substantial)
x = train_df[train_df.ufo2010 < 333]
#calculate mean of population column for no subs. appearances and store it in train_pop_UnSub_mean value
train_pop_UnSub_mean = stats.mean(x['pop'])
#calculate mean of income column for no subs. appearances and store it in train_income_UnSub_mean value
train_income_UnSub_mean = stats.mean(x['income'])
```

```python
#Print mean values
print('mean of population with substantial appearances', train_pop_Sub_mean)
print('mean of income with substantial appearances', train_income_Substantial_mean)
print('mean of population with NO substantial appearances', train_pop_UnSub_mean)
print('mean of income with NO substantial appearances', train_income_UnSub_mean)

#These mean values will show us what the average is for population and incomes for regions
#with and without substantial UFO appearances

#This is simply just a point of reference
```

```python
import math #we will need the math library to carry out computational functions

#This function, predict_appearances, is a training algorithm that classifies
#each region as having substantial UFO appearances according to the 
#closest mean (substantial or not substantial)
#The function takes in two values: the population and the income for that given region
def predict_appearances(pop, income):
    #Subtract datapoints of substantial appearances from their means
    sub_a = train_pop_Sub_mean - pop
    sub_b = train_income_Substantial_mean - income
    
    #Subtract datapoints of non-substantial appearances from their means
    uns_a = train_pop_UnSub_mean - pop
    uns_b = train_income_UnSub_mean - income
    
    #in these two lines below, we will calculate the pythagorean distance
    #these values (sub_dist and uns_dist) calculate the distance between the 
    #population and income of each region to their corresponding means
    #they calculate square root of the sum of the 2 squares
    sub_dist = math.sqrt(math.pow(sub_a, 2) + math.pow(sub_b, 2))
    uns_dist = math.sqrt(math.pow(uns_a, 2) + math.pow(uns_b, 2))
    
    #assign label of the closest mean
    #this makes the output more clear
    print('This region is subject to substantial UFO appearances' if sub_dist < uns_dist 
          else 'This region is NOT subject to substantial UFO appearances')
```

```python
#Here we will test the function by using two different rows from the testing dataframe (test_df)

#first test output should be that the region is not subjected to UFO appearances
predict_appearances(8804, 47240)

#second test output should be that the region is subject to UFO appearances
predict_appearances(9962789, 98148)

#As you can see, the algorithm passed both of these tests since it outputted the correct information
```
